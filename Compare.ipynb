```python
import jiwer
from bert_score import score

original_text = "Văn bản gốc."

whisper_text = "Văn bản của Whisper."

qwen_text = "Văn bản của Qwen."

wer_value = jiwer.wer(original_text, whisper_text)
print(f"--- Kết quả WER (Whisper vs Original) ---")
print(f"Word Error Rate: {wer_value:.4f}")

P, R, F1 = score([qwen_text], [original_text], lang='vi', verbose=False)

print(f"--- Kết quả BERTScore (Qwen vs Original) ---")
print(f"Precision: {P.mean():.4f}")
print(f"Recall: {R.mean():.4f}")
print(f"F1 Score: {F1.mean():.4f}")
```
